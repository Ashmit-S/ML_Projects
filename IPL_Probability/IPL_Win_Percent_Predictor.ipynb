```python
#Group the data based on match id and innings and take the sum of that data i.e find the total runs in each innings of each match
total_scores_df = delivery.groupby(['match_id','inning']).sum()['total_runs'].reset_index()
```

```python
total_scores_df = total_scores_df[total_scores_df['inning'] == 1]
total_scores_df
```

```python
#Merge the total runs of first innings into the match_df
match_df = match.merge(total_scores_df[['match_id','total_runs']],left_on='id',right_on='match_id')
match_df
```

```python
match_df.team1.unique()
```

```python
teams = ['Sunrisers Hyderabad', 
         'Mumbai Indians', 
         'Royal Challengers Bangalore', 
         'Kolkata Knight Riders', 
         'Kings XI Punjab',
         'Chennai Super Kings', 
         'Rajasthan Royals', 
         'Delhi Capitals' ]
```

```python
#Replace older names of teams with newer names, also remove the teams which are now no longer a part of IPL
match_df.team1 = match_df.team1.str.replace('Delhi Daredevils', 'Delhi Capitals')
match_df.team2 = match_df.team2.str.replace('Delhi Daredevils', 'Delhi Capitals')

match_df.team1 = match_df.team1.str.replace('Deccan Chargers', 'Sunrisers Hyderabad')
match_df.team2 = match_df.team2.str.replace('Deccan Chargers', 'Sunrisers Hyderabad')

match_df = match_df[match_df.team1.isin(teams)]
match_df = match_df[match_df.team2.isin(teams)]
```

```python
match_df.shape
```

```python
match_df.dl_applied.value_counts()
```

```python
match_df = match_df[match_df['dl_applied']==0]
```

```python
#Pick only the useful columns from match_df and merge with delivery dataframe to create a new dataframe delivery_df
match_df = match_df[['match_id','city','winner','total_runs']]
delivery_df = match_df.merge(delivery,on='match_id')
delivery_df = delivery_df[delivery_df.inning==2]
```

```python
delivery_df.shape
```

```python
delivery_df.batting_team = delivery_df.batting_team.str.replace('Delhi Daredevils', 'Delhi Capitals')
delivery_df.bowling_team = delivery_df.bowling_team.str.replace('Delhi Daredevils', 'Delhi Capitals')

delivery_df.batting_team = delivery_df.batting_team.str.replace('Deccan Chargers', 'Sunrisers Hyderabad')
delivery_df.bowling_team = delivery_df.bowling_team.str.replace('Deccan Chargers', 'Sunrisers Hyderabad')
```

```python
#Calculate Current Score, Runs left, Balls Left for the second innings
delivery_df['current_score'] = delivery_df.groupby('match_id').cumsum()['total_runs_y']
delivery_df['runs_left'] = delivery_df['total_runs_x'] - delivery_df['current_score'] + 1 
delivery_df['balls_left'] = 126 - (delivery_df['over'] *6 + delivery_df['ball'])
```

```python
#Replace all the NaN values of player dismissed with 0, and 
#Convert all the places where there is a name of the player in the 'player_dismissed' column into 1 and
#Convert the entire column to integer type and then take the cumulative sum to know the number of wickets fallen
#Calculate the wickets remaining
delivery_df['player_dismissed'] = delivery_df['player_dismissed'].fillna("0")
delivery_df['player_dismissed'] = delivery_df['player_dismissed'].apply(lambda x:x if x=="0" else "1")
delivery_df['player_dismissed'] = delivery_df['player_dismissed'].astype('int')
wickets = delivery_df.groupby('match_id').cumsum()['player_dismissed'].values
delivery_df['wickets_left'] = 10 - wickets
delivery_df.tail()
```

```python
#Calculate Current Run Rate and Required Run Rate
delivery_df['current_rr'] = delivery_df['current_score']*6/(120-delivery_df['balls_left'])
delivery_df['required_rr'] = delivery_df['runs_left']*6/delivery_df['balls_left']
```

```python
#If batting team wins, return 1 else return 0
def result(row):
    return 1 if row['batting_team'] == row['winner'] else 0
```

```python
delivery_df['result'] = delivery_df.apply(result,axis=1)
```

```python
final_df = delivery_df[['batting_team','bowling_team', 'city', 'runs_left', 'balls_left', 'wickets_left', 'total_runs_x', 'current_rr', 'required_rr', 'result']]
```

```python
#RandomForest not used because was giving extreme results
```

```python
#To remove a bias we shuffle the dataset
final_df = final_df.sample(final_df.shape[0])
```

```python
final_df.sample()
```

```python
#Remove any NaN values in final_df
final_df.dropna(inplace=True)
```

```python
#To prevent infinite run rate we remove the case when 0 balls remain
final_df = final_df[final_df['balls_left'] != 0]
```

```python
X = final_df.iloc[:,:-1]
Y = final_df.iloc[:,-1]
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.2,random_state=1)
```

```python
X_train
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder

trf = ColumnTransformer([
    ('trf',OneHotEncoder(sparse=False,drop='first'),['batting_team','bowling_team','city'])
]
,remainder='passthrough')
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import Pipeline
```

```python
pipe = Pipeline(steps=[
    ('step1',trf),
    ('step2',LogisticRegression(solver='liblinear'))
])
```

```python
pipe.fit(X_train,Y_train)
```

```python
Y_pred = pipe.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(Y_test,Y_pred)
```

```python
def match_progression(x_df,match_id,pipe):
    #Step 1: Select a particular match by it's "match id" and the end of each over 
    match = x_df[x_df['match_id'] == match_id]
    match = match[(match['ball'] == 6)]
    #Step 2: Make a temperory data frame with all the required columns
    temp_df = match[['batting_team','bowling_team','city','runs_left','balls_left','wickets_left','total_runs_x','current_rr','required_rr']].dropna()
    #Step 3: Remove all the cases where balls_left is 0
    temp_df = temp_df[temp_df['balls_left'] != 0]
    #Step 4: Calculate the probability of losing and winning
    result = pipe.predict_proba(temp_df)
    temp_df['lose'] = np.round(result.T[0]*100,1)
    temp_df['win'] = np.round(result.T[1]*100,1)
    #Step 5: Calculate end_of_over values and other similar values
    temp_df['end_of_over'] = range(1,temp_df.shape[0]+1)
    
    target = temp_df['total_runs_x'].values[0]+1
    runs = list(temp_df['runs_left'].values)
    new_runs = runs[:]
    runs.insert(0,target)
    temp_df['runs_after_over'] = np.array(runs)[:-1] - np.array(new_runs)
    wickets = list(temp_df['wickets_left'].values)
    new_wickets = wickets[:]
    new_wickets.insert(0,10)
    wickets.append(0)
    w = np.array(wickets)
    nw = np.array(new_wickets)
    temp_df['wickets_in_over'] = (nw - w)[0:temp_df.shape[0]]
    
    print("Target-",target)
    temp_df = temp_df[['end_of_over','runs_after_over','wickets_in_over','lose','win']]
    return temp_df,target
    
```
